```python
import numpy as np
from scipy.stats import beta
import torch
import matplotlib.pyplot as plt
import torch.distributions as D
def quadratic(x):
    return 4*x*(1-x)

N = 100000
x = np.empty(N)

# arbitary irrational starting point
x[0] = 1/np.sqrt(3) 
for i in range(1, N):
    x[i] = quadratic( x[i-1] )

plt.plot(x[0:100])
plt.xlabel("iteration index")
plt.show()

t = np.linspace(0, 1, 100)
plt.hist(x, bins=t, normed=True)
plt.xlabel("bins")
plt.ylabel("counts")

```

```python
d8=D.beta.Beta(0.3,0.3)
x=0.2
t=d8.rsample(torch.Size([100000]))
cdf_v1=torch.sum((t<(x-0.01)), dtype=torch.float32)/100000
cdf_v2=torch.sum((t<(x+0.01)), dtype=torch.float32)/100000

cdf_v=(cdf_v2+cdf_v1)/2

d9=D.beta.Beta(0.3,0.3)
d9_sample=np.linspace(0.00001, 0.99999, 10000)
d9_pdf=torch.exp(d9.log_prob(torch.tensor(d9_sample,dtype=torch.float32)))
d9_delta=d9_sample[1]-d9_sample[0]
d9_sum_pdf=torch.sum(d9_pdf*d9_delta)
sum_cdf=0.0;
if(x<0.00001):
    x=0.00002;
if(x>0.99999):
    x=0.99998;
for i in range(len(d9_pdf)):
    if(d9_sample[i]<x):
        sum_cdf+=d9_pdf[i]*d9_delta;
    else:
        sum_cdf+=(d9_pdf[i]+d9_pdf[i-1])/ 2 *((x-d9_sample[i-1]))
        break;
#plt.plot(tt,yy)
#plt.show()
print(cdf_v)
print(sum_cdf/d9_sum_pdf)
```
